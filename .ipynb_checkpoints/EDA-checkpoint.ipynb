```python
# Objective 2: Analyze TB Prevalence and Mortality by Country
Country_stats =data.groupby("Country or territory name")[["Estimated prevalence of TB (all forms) per 100 000 population", "Estimated mortality of TB cases (all forms, excluding HIV) per 100 000 population"
]].mean().dropna()
```

```python
Country_stats
```

```python
 #  Sort countries by highest average prevalence and mortality
top_prevalence = Country_stats.sort_values(by="Estimated prevalence of TB (all forms) per 100 000 population", ascending=False).head(10)
top_mortality = Country_stats.sort_values(by="Estimated mortality of TB cases (all forms, excluding HIV) per 100 000 population", ascending=False).head(10)
 
```

```python
#  Plot side-by-side bar charts
fig, axes = plt.subplots(1, 2, figsize=(18, 6))

# Plot for TB Prevalence
sns.barplot(data=top_prevalence, x="Estimated prevalence of TB (all forms) per 100 000 population", y="Country or territory name", hue="Country or territory name", palette="Reds_r", ax=axes[0])
axes[0].set_title("Top 10 Countries by TB Prevalence (per 100,000)")
axes[0].set_xlabel("Prevalence")
axes[0].set_ylabel("Country")
# Plot for TB Mortality
sns.barplot(data=top_mortality, x="Estimated mortality of TB cases (all forms, excluding HIV) per 100 000 population", y="Country or territory name", hue="Country or territory name", palette="Purples", ax=axes[1])
axes[1].set_title("Top 10 Countries by TB Mortality (per 100,000)")
axes[1].set_xlabel("Mortality")
axes[1].set_ylabel("Country")

plt.tight_layout()
plt.show()
```

```python
import plotly.express as px
import plotly.graph_objects as go
from jupyter_dash import JupyterDash
from dash import Dash, dcc, html, Input, Output
```

```python
data = pd.read_csv("TB_Burden_Country.csv")
data.columns = data.columns.str.strip()  
```

```python
app = Dash(__name__)  
app.title = "TB Burden Dashboard"
```

```python
app.layout = html.Div([
    html.H1("Tuberculosis Burden Dashboard", style={'textAlign': 'center'}),

    html.Div([
        html.Label("Select a Country:"),
        dcc.Dropdown(
            id='country-dropdown',
            options=[{'label': c, 'value': c} for c in sorted(data["Country or territory name"].dropna().unique())],
            value='India'
        )
    ], style={'width': '40%', 'margin': 'auto'}),

    dcc.Graph(id='trend-plot'),
    dcc.Graph(id='scatter-plot'),
    dcc.Graph(id='heatmap')
])
```

```python
@app.callback(
    Output('trend-plot', 'figure'),
    Input('country-dropdown', 'value')
)
def update_line_chart(country):
    filtered = data[data["Country or territory name"] == country]
    fig = go.Figure()
    fig.add_trace(go.Scatter(
        x=filtered["Year"],
        y=filtered["Estimated prevalence of TB (all forms) per 100 000 population"],
        name='Prevalence',
        mode='lines+markers'
    ))
    fig.add_trace(go.Scatter(
        x=filtered["Year"],
        y=filtered["Estimated mortality of TB cases (all forms, excluding HIV) per 100 000 population"],
        name='Mortality',
        mode='lines+markers'
    ))
    fig.update_layout(title=f"TB Trends in {country}", xaxis_title="Year", yaxis_title="Rate per 100,000")
    return fig
```

```python
def update_scatter(_):
    df = data[[
        "Country or territory name",
        "Estimated prevalence of TB (all forms) per 100 000 population",
        "Estimated mortality of TB cases (all forms, excluding HIV) per 100 000 population"
    ]].dropna()

    fig = px.scatter(
        df,
        x="Estimated prevalence of TB (all forms) per 100 000 population",
        y="Estimated mortality of TB cases (all forms, excluding HIV) per 100 000 population",
        text="Country or territory name",
        title="Prevalence vs Mortality by Country"
    )
    fig.update_traces(textposition='top center')
    return fig
```

```python
@app.callback(
    Output('heatmap', 'figure'),
    Input('country-dropdown', 'value')
)
def update_heatmap(_):
    corr = data[[
        "Estimated prevalence of TB (all forms) per 100 000 population",
        "Estimated mortality of TB cases (all forms, excluding HIV) per 100 000 population",
        "Estimated incidence (all forms) per 100 000 population",
        "Case detection rate (all forms), percent"
    ]].dropna().corr()

    fig = px.imshow(
        corr,
        text_auto=True,
        color_continuous_scale='RdBu_r',
        title="Correlation Matrix of TB Metrics"
    )
    return fig
```

```python
if __name__ == "__main__":
    app.run(mode="inline",port=8057)
```
